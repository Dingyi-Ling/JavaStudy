# Ciphers

## Idea for the IEM program
Idea: Interactive Password Security Explorer

Form: A text-based application where users input passwords and choose which cipher they want to use to encode them. The app will visually display how long it would take to decipher them, using different methods such as brute force attacks, dictionary attacks, and algorithm attacks. It aims to demonstrate the complex process behind data encoding and its place in data security.

->When data breaches are becoming more frequent, understanding how passwords are cracked and how encryption works is important. Many people underestimate the vulnerabilities in their passwords, and this program will help users to learn real-world security risks.

-> This program will also teach users on how different encryption techniques impact security and why certain methods works better. By visually showing attack times, users will be easier to understand password strength.

##Suprising Findings
1. Password Length > Complexity - a 12 character password with only one lower case letter could be more secure than an 8 character password with symbols, numbers, abd uppercase letters.

  This challenges the common belief that adding symbols always makes a password stronger. Length is important because longer passwords exponentially increase the number of possible combinations.

2. Cipher Misuse Risks - Even secure ciphers like AES can be vulnerable if implemented incorrectly (using outdated models)
  -> For example, weak key management or poor randomization can render strong encryption useless

3. High Password Reuse Rates: A 2024 survey revealed that 41% of individuals reuse passwords across multiple accounts, with 25% doing so across 11 to 20 sites.
https://www.techopedia.com/password-manager/password-statistics

4. Prevalence of Weak Passwords: Despite awareness campaigns, many users continue to employ easily guessable passwords. Common choices like "123456" and "password" remain prevalent, and such weak passwords can be cracked in less than a second.
